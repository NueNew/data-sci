```
f=open('ChulaEngineering_all_messages.txt')
tokens= f.read().split() # tokens  list 
f.close()

s3=pd.Series(tokens)
s3.value_counts()[:100].plot(kind="bar") #value_counts()  Counter  Collection
```

```
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font = fm.FontProperties(fname='THSarabunNew.ttf')
counts = s3.value_counts()[:100]
ax=counts.plot(kind='bar',figsize=(15,7),color=['b'])
ax.set_xticklabels(counts.index,FontProperties=font)
plt.show()
```
